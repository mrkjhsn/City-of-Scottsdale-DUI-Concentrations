### How are DUI citations distributed across the day?  How are they distributed based on the officers enforcing DUI citations?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pyodbc
import pandas as pd

# databaseConnection = 'DRIVER={SQL Server}; SERVER=ServerName; Database=DatbaseName; UID=UserId; PWD=password;'
connection = pyodbc.connect('DRIVER={SQL Server};SERVER=WINDOWS-A6S01I2\SQLEXPRESS;DATABASE=City_of_Scottsdale;Trusted_Connection=yes;')
cursor = connection.cursor()

# SQL query to import all citation data
Citations = '''
  select *
  from    [dbo].[spd_PDCitations$]  
  '''
Citations = pd.read_sql(Citations, connection)
connection.close()

Citations['Day_of_Week'] = Citations['Citation Date']. \
    apply(lambda x: x.strftime('%a')) #format Citation Day to represent day of week(Mon, Tu, Wed, ect.)

Citations['Hour_of_Day'] = Citations['Citation Time']. \
    apply(lambda x: str(x)[-7:-4]) # extract the hour of the day from the datetime column

Citations['Hour_of_Day'] = pd.to_numeric(Citations['Hour_of_Day'],errors='coerce')

#filter for only citations of type 'DUI'
DUI_Citations = Citations.loc[Citations['Charge Description'].str.contains('DUI')]

#filter for only citations of the type 'DUI-BAC'
DUI_Citations_BAC = Citations.loc[Citations['Charge Description'].str.contains('DUI-BAC')]
```

### Scatterplot of all citations across the day, based on officer badge number of the officer that issued the citation.

#### It's unclear to me why there is a jump in officer badge numbers from about 850 to 1100.  The smaller number of citations issued by the lower badge numbers during late night hours makes me think badges are issued sequentially.  More senior officers will have the ability to select more desirable schedules.

```python
plt.rcParams["figure.figsize"] = [15, 25]

Citations.plot(x = 'Citation Time', y = 'Officer Badge #', kind = 'scatter', color='maroon')
```

### Scatterplot of only DUI citations across the day, based on officer badge number of the officer that issued the citation.

#### My theory that badge numbers are sequential seems to hold up with the below plot.  A larger number of officers with higher badge numbers issued DUI citations more regularly.  However, two more senior officers issued a large number of citations.  Perhaps these officers were working specific DUI task force shifts.  Overall DUI citations are disproportionally enforced by more junior police officers.

```python
plt.rcParams["figure.figsize"] = [15, 25] 

DUI_Citations.plot(x = 'Citation Time', y = 'Officer Badge #', kind = 'scatter', color='maroon')
plt.grid()
```

### Swarmplot of only DUI citations across the day, based on days of the week.

#### Compared to the other days of the week, Sunday evening starting around 8pm has a sharp dropoff, before it baloons out again.

```python
plt.rcParams["figure.figsize"] = [15, 10]    #sets the ploted size of the histogram

sns.swarmplot(x=DUI_Citations['Day_of_Week'], 
                y=DUI_Citations['Citation Time'])
```
